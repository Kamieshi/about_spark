```python
from pyspark.sql import SparkSession
```

```python
session = (
    SparkSession
    .builder
    .master("local[2]")
    .appName("Driver")
    .getOrCreate()
)
```

```python
rdd_set = session.sparkContext.textFile("./input_data.txt")
```

```python
rdd_set.count()
```

```python
phase_one = rdd_set.flatMap(lambda x: x.split(" "))
phase_one.collect()
```

```python
phase_two = phase_one.map(lambda x:(x,1))
phase_two.collect()
```

```python
phase_three =  phase_two.groupByKey()
phase_three.collect()
```

```python
phase_fore = phase_three.map(lambda x:(x[0],len(x[1])))
phase_fore.collect()
```

```python
phase_five = phase_fore.sortBy(lambda x:x[1])
phase_five.collect()
```

```python
easy_solution = rdd_set.flatMap(lambda x: x.split(" ")).map(lambda x: (x, 1)).reduceByKey(lambda x, y: x + y)
easy_solution.sortBy(lambda x:x[1]).collect()
```

```python
phase_five.saveAsTextFile("./output.txt")
```

```python
phase_five.take(10)
```

```python
accum_counter = session.sparkContext.accumulator(0)
```

```python

def add_to(x):
    accum_counter.add(1)
phase_five.foreach(add_to)
accum_counter.value
```
